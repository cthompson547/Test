```python
# 15.1

from time import sleep
from datetime import datetime
import multiprocessing

def process(time):
    sleep(time)
    print(datetime.now())

x = multiprocessing.Process(target=process, args=(1,))

y = multiprocessing.Process(target=process, args=(0,))

z = multiprocessing.Process(target=process, args=(1,))

x.start()
x.join()
y.start()
y.join()
z.start()
z.join()
```

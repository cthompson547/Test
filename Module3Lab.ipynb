```python
class Vehicle:
    def __init__(self, type="misc"):
        self.type = type

class Automobile(Vehicle):
    def __init__(self, year="2016", make="Honda", model="Odyssey", doors="4", roof="Solid"):
        super().__init__("Car")
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

test = Automobile(input("Enter year of vehicle: "), input("Enter brand make: "), input("Enter model: "), input("Enter number of doors: "), input("Enter roof type: "))

print("\VEHICLE INFO\n")
print("Vehicle type:", test.type)
print("Year:", test.year)
print("Make:", test.make)
print("Model:", test.model)
print("Number of doors:", test.doors)
print("Type of roof:", test.roof)
        
```

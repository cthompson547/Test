```python
# 4.1

import random
secret = random.randint(1,10)
try:
    guess = int(input("Input guess:"))
    while guess != "z":
        if guess > 10 or guess < 1:
            raise ValueError("Guess out of range. Range is 1 and 10")
        if guess > secret:
            print("Too high")
            guess = int(input("Input guess:"))
        elif guess < secret:
            print("Too low")
            guess = int(input("Input guess:"))
        else:
            secret = random.randint(1,10)
            print("Correct guess, a new number is generated")
            guess = int(input("Input guess:"))
except Exception as excpt:
    print(f"Exception: {excpt}")
```

```python
# 4.2

boolean = input("Is it small? y/n: ")
if boolean == "y":
    small = True
else:
    small = False
boolean = input("Is it green? y/n: ")
if boolean == "y":
    green = True
else:
    green = False

if small and green:
    print("Pea")
elif not small and green:
    print("Watermelon")
elif small and not green:
    print("Cherry")
else:
    print("Pumpkin")
```

```python
# 6.2

guess_me = 7
number = 1
increment = int(input("Enter an increment"))

while True:
    if number < guess_me:
        print("Too low")
    elif number == guess_me:
        print("Found it")
        break
    else:
        print("Oops")
        break
    number+=increment
```

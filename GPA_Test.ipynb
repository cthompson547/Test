```python
# Module 2 Lab
# Connor Thompson
# GPA Test
# The app will prime the while loop by asking for the students name before entering the loop
# Once the loop is entered, it will ask for the student's GPA
# The app will decide if the person's GPA is high enough for Dean's List and Honor Roll
# If the name is "ZZZ", it will exit the loop

student_name = input("Input student last name")
while student_name != "ZZZ":
    student_GPA = float(input("Enter student GPA"))
    if student_GPA >= 3.5:
        print(f"{student_name} has made it to Dean's List")
    if student_GPA >= 3.25:
        print(f"{student_name} has made it to the Honor Roll")
    student_name = input("Input student last name")
print("End of program")
```
